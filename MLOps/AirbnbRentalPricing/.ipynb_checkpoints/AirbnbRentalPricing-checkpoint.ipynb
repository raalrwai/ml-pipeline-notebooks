```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import xgboost as xgb

```

```python
df = pd.read_csv('AB_NYC_2019.csv')
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
```

```python
print(df.isnull().sum())
```

```python
df = df[(df['price'] > 0) & (df['price'] <= 2000)]
```

```python
df = pd.get_dummies(df, columns=['neighbourhood_group', 'room_type'], drop_first=True)
print(df.columns)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=30, edgecolor='black')
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
top_10_neighborhoods = df.groupby('neighbourhood')['price'].mean().sort_values(ascending=False).head(10)

print(top_10_neighborhoods)
```

```python
numeric_df = df.select_dtypes(include=[np.number])

corr_matrix = numeric_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='plasma', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='room_type_Private room', y='price', data=df)

plt.xticks(ticks=[0, 1], labels=['Other room type', 'Private room'])

plt.title('Average Price vs Room Type')
plt.show()
```

```python
features = ['neighbourhood_group_Brooklyn', 'neighbourhood_group_Manhattan', 
            'neighbourhood_group_Queens', 'neighbourhood_group_Staten Island', 
            'room_type_Private room', 'room_type_Shared room', 
            'number_of_reviews', 'availability_365']

X = df[features]
y = df['price']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

min_max_scaler = MinMaxScaler()
X_minmax_scaled = min_max_scaler.fit_transform(X)
```

```python
df_std = pd.DataFrame(X_scaled, columns=features)
df_minmax = pd.DataFrame(X_minmax_scaled, columns=features)

print("StandardScaler Summary Statistics:")
print(df_std.describe())

print("\nMinMaxScaler Summary Statistics:")
print(df_minmax.describe())
```

```python
print("Comparison of Scaling Methods:")
print("- After applying StandardScaler, all features have mean values close to 0 and standard deviations close to 1, which standardizes features by centering and scaling to unit variance.")
print("- After applying MinMaxScaler, all features are scaled between 0 and 1, preserving the original distribution shape but changing the range.")
print("- For example, the 'number_of_reviews' feature after StandardScaler has mean ~0 and std ~1, while after MinMaxScaler it is bounded between 0 and 1 with a lower standard deviation.")
print("- These differences can affect model training differently, so comparing them helps select the best scaling method for the task.")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "SVR": SVR(),
    "XGBoost": xgb.XGBRegressor()
}
results = {}
```

```python
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    results[model_name] = {'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R²': r2}

results_df = pd.DataFrame(results).T
print(results_df)
```

```python
print("Conclusion:\n"
      "1. Which model performed best?\n"
      "   The Linear Regression and Ridge Regression models performed the best, both achieving a MAE of ~54 and an R² of 0.29. Their performance was consistent and stable across different evaluation metrics.\n"
      "2. Why might that model be more suitable for this dataset?\n"
      "   These models are relatively simple and work well when the relationships between features and the target variable (price) are linear. Given that many factors (like neighborhood or room type) likely affect price in a linear way, these models are a good fit for the data.\n"
      "3. How could you improve predictions further?\n"
      "   - Feature Engineering: Adding new features or transforming existing ones, such as creating interaction terms or including more detailed location data, could help improve performance.\n"
      "   - Outlier Handling: Better handling of extreme price outliers (perhaps by capping them or using different imputation techniques) could help make the model more accurate.\n"
      "   - Advanced Models: Exploring more advanced models like XGBoost or Random Forest could improve predictions by capturing non-linear relationships. Additionally, hyperparameter tuning for these models could help them perform better.")
```
